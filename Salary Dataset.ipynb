```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Salary_dataset.csv',index_col=0)
df.head()
```

```python
sns.scatterplot(df,x='YearsExperience',y='Salary')
plt.plot()
```

```python
sns.kdeplot(df['Salary'])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# X = df.iloc[:,1:2].values
# y = df.iloc[:,-1].values
X = df[['YearsExperience']]   
y = df['Salary']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test.iloc[[0]])
```

```python
lr.predict([[5.3],[3],[2]])
```

```python
lr.predict([[0]])
```

```python
plt.scatter(df["YearsExperience"],df["Salary"])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel("Years Experience")
plt.ylabel("Salary")
```

```python
print("Slope (m):", lr.coef_)
print("Intercept (b):", lr.intercept_)
```

```python
y_pred =  lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print("MSE",mean_squared_error(y_test,y_pred))
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
print("R2 : ",r2 := r2_score(y_test,y_pred))
```

```python
n = X_test.shape[0]   # number of samples
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - 1 - 1)

print("Adjusted R2:", adjusted_r2)
```
